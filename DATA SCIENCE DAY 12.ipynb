## Kurtosis:

                   * Kurtosis measure the shape of the frequency curve
                   * It will measure the degree of the peakness of a frequency distribution
               
          Types of Kurtosis:
                 * Leptokurtic (value > 0 )
                 * Platykurtic (value < 0 )
                 * Mesokurtic  (value == 0)

```r
# install.packages("e1071")
```

```r
library(e1071)
```

```r
x=c(41, 52, 53, 64, 65, 86, 86, 92)
a=kurtosis(x)
print(a)

plot(density(x))
```

```r
# using if and else if condition:

print(paste("The Kurtosis value: ", a))
      
if(a<0){
    print("Platykurtic")
} else if(a>0){
    print("Leptokurtic")
} else if(a==0){
    print("Mesokurtic")
}
```

## T- Test:

                   T-Test will tell the significance difference between the groups
              
                   When to use T-Test:
                          * Data are Independent
                          * Datas are noramlly distributed (check distribution by applying skewness)
                          
                   Types of T-Test:
                          * Paired T-Test (If the groups come from a single population)
                          * Two-Sample T-Test (If the groups come from two different)
                          * One-Sample T-Test (If there is one group being compared against a standard value)
                    
                   T-Test:
                       * One-tailed T-Test
