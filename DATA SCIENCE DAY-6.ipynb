# Reading a Excel file:

## Creating a DataFrame and writing the values in the existing excel file:

```r
df=data.frame(name=c("William","Emma","Nanthiesh","Rohit","Prasanna"), region=c("east","west","north","south","north"),
              sales=c("200000","23980","3848334","893479","93487398"), expenses=c("154824","234873","846785","157863","545312"))
df
```

## Creating a Excel file:
          write.xlsx(df, file="AIMLWRITE.xlsx",sheetName = "Sample")
## for deleting the first column:
          write.xlsx(df, file="AIMLWRITE.xlsx",sheetName = "Sample", row.name = FALSE)
## to read a specific sheet name:
          write.xlsx(df, file="AIMLWRITE.xlsx",sheetName = "New_Titanic" )
## range of row index:
          write.xlsx(df, file="AIMLWRITE.xlsx",sheetName = " New_Titanic", rowIndex=1:10)
## read excel from 3rd row:
          write.xlsx(df, file="AIMLWRITE.xlsx",sheetName = "New_Titanic", startRow=3)
## range of col index:
          write.xlsx(df, file="AIMLWRITE.xlsx",sheetName = "New_Titanic", colIndex=1:3)

```r
# to create another sheet in the existing excel file:
write.xlsx(df,file="mynewAIML.xlsx",sheetName="Exp Data" ,append=TRUE)
```
