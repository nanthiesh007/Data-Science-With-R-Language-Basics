## Various Debugging Functions:

                * Traceback 
                * Browser
                * Recover

 options error = traceback

```r
exponent <- function (num, pow){
    if (pow== 0){
        return(1)
    }
    if (pow==1){
        return (num + "Morfmaert")
        }
    if (pow%% 2 == 1){
        return(num exponent (num, pow - 1))
        } else {
        temp <- exponent (num, floor (pow / 2))
        return(temp temp)
        }
    }
options error traceback
exponent (10, 10)
```

# Regular Expression:

                    Regular expression is a powerful tool for string manipulation
                    
                   * grepl() -> returns the boolean value  
                   * grep()  -> returns the index value
                   * str_extract_all() 

## (grepl) Function:

```r
words = c("python", "java", "Python", "C")
grepl(pattern="Python", x=words,ignore.case=TRUE)
```

```r
words = c("python", "java", "Python", "C")
grep(pattern="Python", x=words,ignore.case=TRUE) # grep will give only the index value which matches the conditon
```

```r
words = c("I love python", "java", "I hate Python", "C")
grep(pattern="Python",words, ignore.case=TRUE, value=TRUE) # value is used to print the whole statement
```

## To Replace a String or a Value:

                 * sub -> replaces value at first only
                 * gsub ->  replaces values in all the places 

```r
df <- "Python is a collaborative project with many contributors-R and R is a high level language"
sub('R','R language',df)
```

```r
df <- "Python is a collaborative project with many contributors-R and R is a high level language"
gsub('R','R language',df) #(old,new,input)
```
